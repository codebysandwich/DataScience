### ÂÖ∑‰ΩìÁ§∫‰æã(Ê∞îË±°Êï∞ÊçÆ‰∏ãËΩΩ)

ËøõË°åÊ∞îË±°Êï∞ÊçÆ‰∏ãËΩΩÔºå‰ΩøÁî®ÂáΩÊï∞ËøõË°åÂ∞ÅË£ÖÔºõÊ∞îË±°Êï∞ÊçÆ‰∏ãËΩΩÁ§∫‰æã[weather download](https://github.com/codebysandwich/WorkSpace/blob/main/PyScript/WeatherDownload/app.py)

ÁΩëÁ´ôÁ§∫‰æãhttps://q-weather.info/weather/58354/history/?date=2020-07-01

````python
import pandas as pd

# ÁΩëÁ´ôÁ§∫‰æãËøõË°åÊï∞ÊçÆ‰∏ãËΩΩ
url = r'https://q-weather.info/weather/58354/history/?date=2020-07-01'
pd.read_html(url)[0]
````

ËøôÈáåÁ´ôÁÇπ‰ø°ÊÅØÂíåÊó•ÊúüÊòØÂèØÂèòÁöÑÔºåÊåâÈúÄË∞ÉÊï¥„ÄÇ‰∏ÄËà¨Êàë‰ª¨ÈúÄË¶Å‰∏ãËΩΩ‰∏ÄÊÆµÊó∂Èó¥ÁöÑÊï∞ÊçÆÔºåÊåâÁ´ôÁÇπËøõË°åÊï∞ÊçÆÊ±áÊÄªÔºö

ÈÇ£‰πàÊàë‰ª¨Â∫îËØ•ËÆæËÆ°‰∏Ä‰∏™ÂáΩÊï∞ÔºöÊé•Âèó‰∏§‰∏™ÂèÇÊï∞ÔºàÁ´ôÁÇπÂàóË°®ÔºåÊó∂Èó¥ÂàóË°®ÔºâÔºåÂØπÊØè‰∏™Á´ôÁÇπÊåâÊó∂Èó¥ÂàóË°®ËøõË°å‰∏ãËΩΩÔºåÂπ∂ÊåâÁ´ôÁÇπÂ≠òÂ•ΩxlsxÊñá‰ª∂

````python
def weather_download(stations: list, date_range: list):
    for station in stations:
        # ÊåâÁ´ôÁÇπËøõË°åÊï∞ÊçÆÂ≠òÂÇ®
        station_data = []
        for date in date_range:
            url = r'https://q-weather.info/weather/{0}/history/?date={1}'.format(station, date)
            # ÊØè‰∏™Êó•ÊúüÊï∞ÊçÆËøõË°å‰∏ãËΩΩ
            station_data.append(pd.read_html(url)[0])
            # ÁÆÄÂçïÊòæÁ§∫‰∏ãËøõÂ∫¶
            print(f"{station}: {date} ‰∏ãËΩΩÂÆåÊàêÔºÅ")
        # Á´ôÁÇπÊãºÊé•Êï∞ÊçÆ
        pd.concat(station_data).to_excel('{}Á´ôÁÇπ{}.xlsx'.format(station, 
                                                              '{}-{}'.format(date_range[0], date_range[-1])),
                                        index=None)
````

````python
stations = ['58354', '54398']
date_range = [dt.strftime("%Y-%m-%d") for dt in pd.date_range('2021-10-1', '2021-10-07')]
weather_download(stations, date_range)
````

````python
from PIL import Image
# ÂõæÂÉèÁöÑÊâãÁªò
"""ÈªëÁôΩÈ£éÊ†º
   ËæπÁïåÁöÑ‰ΩçÁΩÆÊØîËæÉÈáç
   Áõ∏ÂêåÊàñÁõ∏ËøëËâ≤ÂΩ©Ë∂ãËøë‰∫éÁôΩËâ≤
   Áï•ÊúâÂÖâÊ∫êÊïàÊûú
"""
# ËØªÂèñÂΩ©Ëâ≤ÂõæÁâáÂπ∂ËΩ¨Âåñ‰∏∫npÊï∞ÁªÑ
a = np.array(Image.open(r'../assets/pics/yasuo.jpg').convert('L')).astype('float')

depth = 10.
grad = np.gradient(a)
grad_x, grad_y = grad
grad_x = grad_x*depth/100
grad_y = grad_y*depth/100
A = np.sqrt(grad_x**2 + grad_y**2 + 1.)
uni_x = grad_x/A
uni_y = grad_y/A
uni_z = 1./A

vec_el = np.pi/2.2
vec_ez = np.pi/4.
dx = np.cos(vec_el)*np.cos(vec_ez)
dy = np.cos(vec_el)*np.sin(vec_ez)
dz = np.sin(vec_el)

b = 255*(dx*uni_x + dy*uni_y + dz*uni_z)
b = b.clip(0, 255)

im = Image.fromarray(b.astype('uint8'))
# ‰øùÂ≠òËΩ¨ÂåñÂêéÁöÑÂõæÁâá
im.save('yasuo_hand.jpg')
````
