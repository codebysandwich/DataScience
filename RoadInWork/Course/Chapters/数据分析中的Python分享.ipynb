### 具体示例(气象数据下载)

进行气象数据下载，使用函数进行封装；气象数据下载示例[weather download](https://github.com/codebysandwich/WorkSpace/blob/main/PyScript/WeatherDownload/app.py)

网站示例https://q-weather.info/weather/58354/history/?date=2020-07-01

````python
import pandas as pd

# 网站示例进行数据下载
url = r'https://q-weather.info/weather/58354/history/?date=2020-07-01'
pd.read_html(url)[0]
````

这里站点信息和日期是可变的，按需调整。一般我们需要下载一段时间的数据，按站点进行数据汇总：

那么我们应该设计一个函数：接受两个参数（站点列表，时间列表），对每个站点按时间列表进行下载，并按站点存好xlsx文件

````python
def weather_download(stations: list, date_range: list):
    for station in stations:
        # 按站点进行数据存储
        station_data = []
        for date in date_range:
            url = r'https://q-weather.info/weather/{0}/history/?date={1}'.format(station, date)
            # 每个日期数据进行下载
            station_data.append(pd.read_html(url)[0])
            # 简单显示下进度
            print(f"{station}: {date} 下载完成！")
        # 站点拼接数据
        pd.concat(station_data).to_excel('{}站点{}.xlsx'.format(station, 
                                                              '{}-{}'.format(date_range[0], date_range[-1])),
                                        index=None)
````

````python
stations = ['58354', '54398']
date_range = [dt.strftime("%Y-%m-%d") for dt in pd.date_range('2021-10-1', '2021-10-07')]
weather_download(stations, date_range)
````

#### 绘制雷达图

````python
'''设置允许中文绘图'''
plt.rcParams['font.family'] = 'PingFang HK'
# ---------------------------------------------

fig = plt.figure(figsize=(6, 6), dpi=120)
ax = fig.add_subplot(111, polar=True)
# 先绘制上限值
values = ser[5:10].to_list()   # 上限值
labels = ['$SO_2$', '$NO_2$', '$PM_{10}$', '$CO$', '$PM_{2.5}$']
# 定义颜色
line_color = np.array([215, 215, 215])/255
n = len(values)
angles = np.concatenate((np.linspace(0, 2*np.pi, n, endpoint=False), [0]))
_max = max(values)
for ang in angles:
    ax.plot([ang]*2, [0, _max], '-', c=line_color)
# -------绘制边界------- #
tick = 3
for i in range(tick):
    ax.plot(angles, [_max*(1/tick)*(i+1)]*(n+1), c = line_color)
# -------绘制数据------- #
ax.plot(angles, values+[values[0]], "--", c = 'r', label='上限值')
# 绘制标准值
values = [1] * 6
ax.plot(angles, values, '-', c='k', label='标准值')
# 绘制特征值
values = ser[:5].to_list()
ax.plot(angles, values+[values[0]], '-', c='orange', label='特征值')
# 绘制下限值
values = ser[10:].to_list()
ax.plot(angles, values+[values[0]], '--', c='green', label='下限值')

ax.set_thetagrids(angles * 180 / np.pi, labels) # 设置显示的角度，将弧度转换为角度

ax.set_theta_zero_location('N')        # 设置极坐标的起点（即0°）在正北方向，即相当于坐标轴逆时针旋转90°
ax.spines['polar'].set_visible(False)  # 不显示极坐标最外圈的圆
ax.grid(False)                         # 不显示默认的分割线
ax.set_yticks([])                      # 不显示坐标间隔
ax.legend()
plt.show()
````

## 做一些好玩的

图片手绘风格转化

````python
from PIL import Image
# 图像的手绘
"""黑白风格
   边界的位置比较重
   相同或相近色彩趋近于白色
   略有光源效果
"""
# 读取彩色图片并转化为np数组
a = np.array(Image.open(r'../assets/pics/yasuo.jpg').convert('L')).astype('float')

depth = 10.
grad = np.gradient(a)
grad_x, grad_y = grad
grad_x = grad_x*depth/100
grad_y = grad_y*depth/100
A = np.sqrt(grad_x**2 + grad_y**2 + 1.)
uni_x = grad_x/A
uni_y = grad_y/A
uni_z = 1./A

vec_el = np.pi/2.2
vec_ez = np.pi/4.
dx = np.cos(vec_el)*np.cos(vec_ez)
dy = np.cos(vec_el)*np.sin(vec_ez)
dz = np.sin(vec_el)

b = 255*(dx*uni_x + dy*uni_y + dz*uni_z)
b = b.clip(0, 255)

im = Image.fromarray(b.astype('uint8'))
# 保存转化后的图片
im.save('yasuo_hand.jpg')
````
