[官方参考文档](https://matplotlib.org/tutorials/colors/colormap-manipulation.html#sphx-glr-tutorials-colors-colormap-manipulation-py)给出了自定义的几种类型。

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
def plot_examples(colormaps):
    """
    Helper function to plot data with associated colormap.
    """
    np.random.seed(19680801)
    data = np.random.randn(30, 30)
    n = len(colormaps)
    fig, axs = plt.subplots(1, n, figsize=(n * 2 + 2, 3),
                            constrained_layout=True, squeeze=False)
    for [ax, cmap] in zip(axs.flat, colormaps):
        psm = ax.pcolormesh(data, cmap=cmap, rasterized=True, vmin=-4, vmax=4)
        fig.colorbar(psm, ax=ax)
    plt.show()
```

```python
hsv = cm.get_cmap('hsv', 256)
newcolors = hsv(np.linspace(0, 1, 300))
newmap = ListedColormap(newcolors[:150])
plot_examples([cm.get_cmap('hsv', 256), newmap])
```

```python
colors = ['#024CEB', '#02BBA9', '#65FF00', '#FEFF00', '#FF8800', '#D40608']
cmap = LinearSegmentedColormap.from_list('mymap', colors)
nodes = [0, .2, .4, .6, .8, 1.]
cmap1 = LinearSegmentedColormap.from_list('mymap', list(zip(nodes, colors)))
plot_examples([cmap, cmap1])
```
