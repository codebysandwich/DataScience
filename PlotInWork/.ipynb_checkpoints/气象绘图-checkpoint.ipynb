## cartopy+matplotlib绘制全球温度分布

### 准备数据

数据下载地址：[https://crudata.uea.ac.uk/cru/data/hrg/cru_ts_4.03/cruts.1905011326.v4.03/tmp/](https://crudata.uea.ac.uk/cru/data/hrg/cru_ts_4.03/cruts.1905011326.v4.03/tmp/)<br>
数据比较大，大概2.95G自备梯子下载或联系作者。

### 数据处理

```python
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset
import cartopy.crs as ccrs
```

```python
filename = r'/Users/caizhiming/Documents/MyCode/TO_YY/files/cru_ts4.03.1901.2018.tmp.dat.nc'
data = Dataset(filename)
```

```python
data.variables.keys()
```

+ 处理温度数据

```python
temp = data.variables['tmp'][:]
temp_av_1901_2018 = np.mean(temp[:, :, :], axis=0)
lat = data.variables['lat'][:]
lon = data.variables['lon']
```

### 绘图

```python
plt.figure(figsize=(13,6.2))
ax = plt.subplot(111, projection=ccrs.PlateCarree())
mm = ax.pcolormesh(lon, lat, temp_av_1901_2018, vmin=-45, 
                   vmax=45, transform=ccrs.PlateCarree(),
                   cmap='nipy_spectral')
ax.coastlines()
plt.colorbar(mm, ax=ax, orientation='vertical')
plt.show()
```

```python
plt.figure(figsize=(13,6.2), dpi=120)
ax = plt.subplot(111, projection=ccrs.PlateCarree())
mm = ax.pcolormesh(lon, lat, temp_av_1901_2018, vmin=-45, 
                   vmax=45, transform=ccrs.PlateCarree(),
                   cmap='nipy_spectral')
ax.coastlines()
plt.colorbar(mm, ax=ax, orientation='vertical', pad=.03, shrink=.89)
plt.title('1901-2018 Global average temperature')
plt.show()
```
